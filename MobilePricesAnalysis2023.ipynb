# Q2)Which phonetype has the highest average rating, and which one has the lowest?

```python
average_ratings_by_brand = df.groupby('PhoneBrand')['Avg_Ratings/5'].mean().sort_values(ascending=False)

highest_rated_brand = average_ratings_by_brand.idxmax()
highest_average_rating = average_ratings_by_brand.max()

lowest_rated_brand = average_ratings_by_brand.idxmin()
lowest_average_rating = average_ratings_by_brand.min()

print(f"The brand with the highest average rating is {highest_rated_brand} with an average rating of {highest_average_rating:.2f}")
print(f"The brand with the lowest average rating is {lowest_rated_brand} with an average rating of {lowest_average_rating:.2f}")
```

# Q12)Is there a relationship between the processor Type and the average rating of the phones?

```python
from scipy import stats

# Assuming your dataset is stored in a DataFrame called 'df'
# Group the data by 'ProcessorType' and calculate the average rating
processor_avg_ratings = df.groupby('ProcessorType')['Avg_Ratings/5'].mean().reset_index()

# Perform ANOVA to test for differences in average ratings between processor types
f_statistic, p_value = stats.f_oneway(*[group['Avg_Ratings/5'] for name, group in df.groupby('ProcessorType')])

print(f"ANOVA F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Create a bar plot to visualize average ratings by processor type
plt.figure(figsize=(10, 6))
sns.barplot(data=processor_avg_ratings, x='ProcessorType', y='Avg_Ratings/5')
plt.title("Average Ratings by Processor Type")
plt.xlabel("Processor Type")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()
```
